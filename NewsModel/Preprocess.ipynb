```python
# 샘플 파일 하나 읽기
sample = pd.read_csv("./2017Samsung/2017-01-01Samsung.csv")
```

```python
sample.columns
```

```python
news = sample['본문'].tolist() # 뉴스본문만 별도 리스트로
```

```python
front_slice = news[0].split("댓글")[1] #split의 앞 잘라내고 뒤 저장
back_slice = front_slice.split("@")[0] #split의 뒤 잘라내고 앞 저장
back_slice
```

```python
from konlpy.tag import Kkma
from konlpy.utils import pprint
```

```python
okt = Okt()
raw = set(okt.pos(back_slice)) #(단어, 형태소) 튜플로
raw_list = list(raw)
len(raw_list)
```

```python
raw_list # 지울 것!!
```

```python
removable_form = ['Josa', 'Punctuation', 'Foreign','Number', 'Suffix', 'Modifier']
removable_word = ['것', '이', '그', '저', '이것', '저것', '그것', '여기', '저기', '거기', '그랬다','됐고', '됐다', '되었다', '됐던', '된',
'라며', '라고', '하며', '하고', '이다', '이고', '이라고', '였다', '이었다', '하고', '하는', '한', '기자', '는', '를']

for tuples in raw_list:
  form = tuples[1]
  word = tuples[0]
  if form in removable_form or word in removable_word:
    raw_list.remove(tuples)

filtered_list = raw_list
len(raw_list)
```
