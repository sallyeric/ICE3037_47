```python
import scipy as sp
import pandas as pd
import numpy as np
import os
import re
from tqdm import tqdm
```

```python
# Dataframe 새로 생성
df_map = pd.DataFrame(index=range(0,1461), columns=['date', 'news'])
```

```python
# 뉴스데이터 가공 후 df_map에 넣음
def get_data():
    date_index = pd.date_range(start='20170101', end='20201231')
    date_list = date_index.strftime("%Y-%m-%d").tolist()

    i=0
    for date in date_list:
        df_next = pd.read_csv('./SamsungNews/%sSamsung.csv'%date)
        daily_result = daily_news(date, df_next)
        df_map['date'][i] = date
        df_map['news'][i] = daily_result
        i+=1

get_data()
df_map
```

```python
# 뉴스 필터링 후 같은 일자마다 이어붙임
def daily_news(date, df_next):
  daily_csv = df_next

  news_data = ""
  for i in tqdm(range(len(df_next))):
    #check empty
    if len(df_next.loc[:,'본문'][i]) == 2:
        continue
    data = df_next.loc[:,'본문'][i].split("댓글")[1]
    data = data.split("@")[0].strip()
    content_text = re.sub(r'\([^)]*\)', '', data)
    news_data += content_text

  return news_data
```

```python
# 뉴스가 일자별로 잘 모였는지 확인
# df_map['news'][1460]
```

```python
# 차트데이터 불러와서 날짜 형식 변경

chart = pd.read_csv('samsung.csv')
chart_date_list = chart['일자']
print(chart['일자'][0])

i=0
for chart_date in chart_date_list:
    str_date = str(chart_date)
    chart['일자'][i] = str_date[0:4] + '-' + str_date[4:6] + '-' + str_date[6:8]
    i+=1
    
chart
```

```python
# 날짜가 뒤집어져 있어서 역순 재배치
chart_reverse = chart.loc[::-1].reset_index(drop=True)
chart_reverse
```

```python
# 20170101, 20201231 데이터 임의로 추가
data20170101 = {'일자':'2017-01-01', '시가':'0', '고가':'0', '저가':'0', '종가':'0', '거래량':'0','변화량':'#VALUE!'}
idx = 0
temp1 = chart_reverse[chart_reverse.index < idx]
temp2 = chart_reverse[chart_reverse.index >= idx]
chart_reverse = temp1.append(data20170101,ignore_index=True).append(temp2, ignore_index=True)

data20201231 = {'일자':'2020-12-31', '시가':'0', '고가':'0', '저가':'0', '종가':'0', '거래량':'0','변화량':'#VALUE!'}
chart_reverse = chart_reverse.append(data20201231,ignore_index=True)

chart_reverse
```

```python
# Column명 date로 통일
chart_reverse.rename(columns = {'일자' : 'date'}, inplace = True)
chart_reverse
```

```python
# 두 개 데이터프레임 합치기
merge = pd.merge(df_map, chart_reverse, how='left', on='date')
merge.tail(10)
```

```python
# NaN 처리 편하게 하려고 잠시 역순으로 뒤집음
merge_reverse = merge.loc[::-1].reset_index(drop=True)
merge_reverse.head(5)
```

```python
# NaN 값을 임의로 99.9로 지정
merge_reverse = merge_reverse.fillna(99.9)
merge_reverse.head(10)
```

```python
# 주말과 휴일 등, 등락수치 없는 경우(NaN, 99.9)
# 그 이후 가장 최근의 주가수치로 통일
    
# 시가  고가  저가  종가  거래량  변화량
for i in range(0,1460):
    if merge_reverse['변화량'][i] == 99.9:
        merge_reverse['시가'][i] = merge_reverse['시가'][i-1]
        merge_reverse['고가'][i] = merge_reverse['고가'][i-1]
        merge_reverse['저가'][i] = merge_reverse['저가'][i-1]
        merge_reverse['종가'][i] = merge_reverse['종가'][i-1]
        merge_reverse['거래량'][i] = merge_reverse['거래량'][i-1]
        merge_reverse['변화량'][i] = merge_reverse['변화량'][i-1]
        
merge_reverse.head(10)       
```

```python
# 다시 역순으로 (최종)
df_final = merge_reverse.loc[::-1].reset_index(drop=True)
df_final
```

```python
# csv 저장
df_final.to_csv("SamsungMapped.csv", mode='a', header=False)
```
